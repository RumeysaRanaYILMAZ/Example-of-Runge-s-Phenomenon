# Example of Runge's Phenomenon

```python
 import numpy as np
 from scipy import interpolate
 import matplotlib.pyplot as plt
```

```python
A=np.zeros([11,2])
A[:,1]=np.arange(0,1.1,0.1)
A[:,0]=np.sin(A[:,1]*np.pi*4)*np.sin(A[:,1]*np.pi*4)
```

```python
plt.plot(A[:,1],A[:,0], 'o', label="Messwerte", c='k')
```

```python
p10=interpolate.lagrange(A[:,1],A[:,0])
x=np.arange(0,1,0.001)
y=p10(x)


plt.plot(A[:,1],A[:,0], 'o',label='Messwerte')

plt.plot(x, y, label="Polynome Ordnung 10", linestyle='-'
, c='k')

B1=np.arange(0,1.1,0.01)
B2=np.sin(B1*np.pi*4)*np.sin(B1*np.pi*4)
plt.plot(B1, B2, label="sin^2(4pix)", linestyle='--', c='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

In this example, we see that it is not useful to interpolate the function at high interpolation degrees. That's what the Rudge phenomenon says. Although the margin of error is low in the middle parts, we see that interpolation between nodes has become worthless in the border parts.

# Adding New Support Points To The Boudary

Let's move the support points closer to the boundary without changing the number of support points.


```python
A=np.zeros([11,2])
A[:,1]=np.array([0, 0.05, 0.1, 0.18, 0.4, 0.5, 0.6, 0.81, 0.9, 0.95, 1])
A[:,0]=np.sin(A[:,1]*np.pi*4)*np.sin(A[:,1]*np.pi*4)


plt.plot(A[:,1],A[:,0], 'o', label="Messwerte", c='k')

```

```python
p10=interpolate.lagrange(A[:,1],A[:,0])
x=np.arange(0,1,0.001)
y=p10(x)

plt.plot(A[:,1],A[:,0], 'o',label='Messwerte')

plt.plot(x, y, label="Polynome Ordnung 10", linestyle='-'
, c='k')

B1=np.arange(0,1.1,0.01)
B2=np.sin(B1*np.pi*4)*np.sin(B1*np.pi*4)
plt.plot(B1, B2, label="sin^2(4pix)", linestyle='--', c='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

As we can see this time around, we can predict with less error in the border and in the middle. But in other parts, we observe that the error has increased. In other words, we have made worse predictions than the first interpolation estimation.
